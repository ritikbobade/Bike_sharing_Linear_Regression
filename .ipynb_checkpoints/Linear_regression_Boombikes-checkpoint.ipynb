In the dataset, we identified six categorical variables.

To investigate their impact on the dependent variable ('cnt'), we employed Box plots (refer to the figure above). Our analysis yielded the following insights:

Season: Approximately 32% of bike bookings occurred during season 3, with a median of over 5000 bookings over a 2-year period. This was followed by season 2 and season 4, which accounted for 27% and 25% of total bookings, respectively. These findings suggest that the season could be a valuable predictor for the dependent variable.

Month (mnth): About 10% of bike bookings took place in months 5, 6, 7, 8, and 9, each with a median of over 4000 bookings per month. This suggests that there is a discernible trend in monthly bookings, making 'mnth' a potential predictor for the dependent variable.

Weather Situation (weathersit): Around 67% of bike bookings occurred during 'weathersit1', with a median close to 5000 bookings over a 2-year period. 'Weathersit2' accounted for 30% of total bookings. These results imply that 'weathersit' exhibits a discernible trend related to bike bookings and could serve as a useful predictor for the dependent variable.

Holiday: A striking 97.6% of bike bookings happened on non-holiday days, indicating a clear bias in the data. This suggests that 'holiday' is unlikely to be a meaningful predictor for the dependent variable.

Weekday: The 'weekday' variable displays a consistent trend, with bookings ranging from 13.5% to 14.8% of the total on all days of the week. The independent medians for each day fall between 4000 and 5000 bookings. Whether this variable should be included in the model as a predictor or not can be determined by the model itself, as it may have some influence, but its significance remains uncertain.

## Creating dummy variables

```python
df[['season','weathersit','month']] = df[['season','weathersit','month']].astype('category')
```

```python
# df['month']=df.month.astype(object)
# df['season']=df.season.astype(object)
# df['weathersit']=df.weathersit.astype(object)
# df['weekday']=df.weekday.astype(object)
# df.info()
```

#### Creating dummy variables for categorical variables

```python
Season_condition=pd.get_dummies(df['season'],drop_first=True)
Weather_condition=pd.get_dummies(df['weathersit'],drop_first=True)
Day_of_week=pd.get_dummies(df['weekday'],drop_first=True)
Month=pd.get_dummies(df['month'],drop_first=True)
```

```python
df=pd.concat([df,Season_condition],axis=1)
df=pd.concat([df,Weather_condition],axis=1)
df=pd.concat([df,Day_of_week],axis=1)
df=pd.concat([df,Month],axis=1)
df.info()
```

```python
#Deleting the orginal columns season.weathersit,weekday,mnth
# df.drop(['season'],axis=1,inplace=True)
df.drop(['weathersit'],axis=1,inplace=True)

df.drop(['weekday'],axis=1,inplace=True)

df.drop(['mnth'],axis=1,inplace=True)

df.head()
```

### Train Test Split

```python
# Now our data is ready to split 
from sklearn.model_selection import train_test_split

# We should specify 'random_state' so that the train and test data set always have the same rows, respectively

np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.70, test_size = 0.30, random_state = 333)
```

### Building a Linear Model
Dividing into X and Y sets for the model building

```python
# df_train.columns=df_train.columns.astype(str)
```

```python
y_train = df_train.pop('count')
X_train = df_train
```

RFE
Recursive feature elimination: We will be using the LinearRegression function from SciKit Learn for its compatibility with RFE (which is a utility from sklearn)

```python
# Importing RFE and LinearRegression
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
# X_train=X_train.astype(float)
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```
